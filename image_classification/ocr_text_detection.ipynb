```python
from paddleocr import PaddleOCR
import cv2
from paddleocr import TextDetection

model = TextDetection(model_name="PP-OCRv5_server_det")
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
import os
import cv2

folder_dir = "C:/Users/Rakesh Kumar/VSCode/Capstone/datasets/medical_drug_images/Salbutamol"

# Output folder
output_dir = "cropped_words"
os.makedirs(output_dir, exist_ok=True)

count = 0

for image_name in os.listdir(folder_dir):

    # Filter image files
    if image_name.lower().endswith((".png", ".jpg", ".jpeg")):

        # Full image path
        img_path = os.path.join(folder_dir, image_name)
        print(f"Processing: {img_path}")

        # Load image
        img = cv2.imread(img_path)
        if img is None:
            print(f"Skipping (failed load): {img_path}")
            continue

        # Run prediction
        output = model.predict(img_path, batch_size=1)

        for res in output:
            res.print()
            res.save_to_img(save_path="./output/")

            dt_polys = res['dt_polys']

            for box in dt_polys:
                x_coords = box[:, 0]
                y_coords = box[:, 1]

                x_min, x_max = int(x_coords.min()), int(x_coords.max())
                y_min, y_max = int(y_coords.min()), int(y_coords.max())

                # Padding
                pad = 5
                x_min = max(0, x_min - pad)
                y_min = max(0, y_min - pad)
                x_max = min(img.shape[1], x_max + pad)
                y_max = min(img.shape[0], y_max + pad)

                crop = img[y_min:y_max, x_min:x_max]

                # Save cropped word
                save_path = os.path.join(output_dir, f"word_{count}.jpg")
                cv2.imwrite(save_path, crop)

                count += 1
```
